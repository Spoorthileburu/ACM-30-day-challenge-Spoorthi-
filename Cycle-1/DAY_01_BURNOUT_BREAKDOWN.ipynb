```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('mental_health_workplace_survey.csv')
print(df.head())
print("missing values in this data set is: ")
print(df.isnull().sum())
missing = df.isnull().sum().sum()
print("No.of missing values in this data set is: ", missing)
```

First, we import pandas as we need it for data analysis and also seaborn and matplot.lib, so as to plot graph or to visualize relationships.Then,we check if there are any missing values using- df.isnull().sum(). if there are any missing values we will substitute mean, mode or median depending on the situation. But as we can see from the output no data is missing.

```python
#checking for outliners
print("Checking for outliners: ")
plt.title('Boxplot of WorkHoursPerWeek')
sb.boxplot(x=df['WorkHoursPerWeek'])
plt.show()
plt.title('Boxplot of StressLevel')
sb.boxplot(x=df['StressLevel'])
plt.show()
plt.title('Boxplot of SleepHours')
sb.boxplot(x=df['SleepHours'])
plt.show()
def find_outliers(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lb = q1 - 1.5 * iqr
    ub = q3 + 1.5 * iqr
    outliers = column[(column < lb) | (column > ub)]
    return outliers

# Checking numerically for outliers
print("\n Outlier:")
for y in ['WorkHoursPerWeek', 'StressLevel', 'SleepHours']:
    outliers = find_outliers(df[y])
    print(f"{y}: {len(outliers)} outliers")
```

Here, we have checked for outliers.

```python
#Let's visualize.......using line graph
plt.figure(figsize=(14,5))
plt.title('The line graph showing stresslevel vs sleephours')
sb.lineplot(x=df['StressLevel'],y=df['SleepHours'])
plt.figure(figsize=(14,5))
plt.title('The line graph between mental health vs burnout')
sb.lineplot(x=df['MentalHealthDaysOff'],y=df['BurnoutLevel'])
#let's use another type of plotting method
plt.figure(figsize=(14,3))
plt.title('The bar graph of stresslevel vs sleephours')
sb.barplot(x=df['StressLevel'],y=df['SleepHours'])
plt.figure(figsize=(10,3))
plt.title('The bar graph of mentalhealthdaysoff vs burnout level')
sb.barplot(x=df['MentalHealthDaysOff'],y=df['BurnoutLevel'])
```

we are using different typeof plotting methods to visualize the data given.
we can see the relation between them.
